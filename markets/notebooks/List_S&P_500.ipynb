# Daily Information each ticker

```python
l_tickers_df = []
```

```python
for ticker in tickers:
    if ticker != 'BF.B' or ticker != 'BRK.B':
        data = yf.download(ticker, start=start_date, end=end_date, interval="1d")
        l_tickers_df.append(data)
```
